MLP structure and training details (using AdamW optimizer) for the datasets

```python
import sys

sys.path.insert(0, "/home/liu00980/Documents/multimodal/tabular/tab-ddpm")

import os
import lib

import pandas as pd
```

```python
def get_inf(ds_name = "abalone"):
    config_dict = lib.load_config(os.path.join(ds_name, "ddpm_cb_best/config.toml"))
    
    training_config = {
        "ds_name": ds_name,
        "hidden_dim": str(config_dict["model_params"]["rtdl_params"]["d_layers"]),
        "diffusion_timestamps": config_dict["diffusion_params"]["num_timesteps"],
        "training_iterations": config_dict["train"]["main"]["steps"],
        "lr": config_dict["train"]["main"]["lr"],
        "batch_size": config_dict["train"]["main"]["batch_size"],
        "weight_decay": config_dict["train"]["main"]["weight_decay"],
    }
    
    return training_config
```

```python
ds_name_list = [
    "fb-comments",
    "adult",
    "house",
    "california",
    "churn2",
    "gesture",
    "abalone",
    "insurance",
]

config_list = [get_inf(ds_name) for ds_name in ds_name_list]

df_dict = {key: [config[key] for config in config_list] for key in config_list[0].keys()}

df = pd.DataFrame(df_dict)
```
